```python
#
# Encapsulation 
#  (wrap code and data into a single object and restrict access to it)
#
# Abstraction 
#  (hide implementation details)  sum() len() capitalize() etc
#   abstract class Vehicle abstract methods start/stop
#   concrete classes Car(Vehicle): Bike(Vehicle): and methods start/stop
#
# Private & Public vars
#
# Inherit child inherits from parent
#

class PlayerCharacter:

    # class atttributes are static
    membership = True

    # Dunder (double underscore) method __init__  class constructor called every time object instantiated 
    # self refers to the class
    def __init__(self, name, sex, age=15):
        if (age < 18 ):
            self.membership = False
        self.name = name                 
        self._age = age   # single _underscore convention denotes private but not enforced    
        self._sex = sex   # single _underscore convention denotes private but not enforced    
    
    def run(self):
        return self
    
    def speak(self):
        print(f'my name is {self.name} i am {self._sex} and {self._age} years old')
    
# classmethod sum_nums, which creates cls PlayerCharacter assigns to and instantiates object player3
teddyBoy = PlayerCharacter('Teddy', 'male', 23)

# calls run method
teddyBoy.run()
teddyBoy.speak()

# works because .name is not ._name private
teddyBoy.name = teddyBoy.name + ' Bear'
teddyBoy.speak()

msg = 'hello'
print(str(msg.capitalize()))
```

```python
class PlayerCharacter:

    # class atttributes are static
    membership = True

    # Dunder (double underscore) method __init__  class constructor called every time object instantiated 
    # self refers to the class
    def __init__(self, name, sex, age=15):
        if (age < 18 ):
            self.membership = False
        self.name = name           
        self.age = age
        self.sex = sex
    
    def run(self, player):
        print(f'{player}: {self.name}, {self.age}, {self.sex}, min-age membership {self.membership}')
    
    # classmethod decorator is bound to the class i.e can be called without instantiating an object  
    @classmethod
    def my_class_method(cls, name, num1, num2, sex):    # special 1st argument 'cls' refers to itself, i.e takes itself as 1st arg

        # method attribute
        nums = [num1, num2]
        print(f'my_class_method membership: {cls.membership} instantiating class {cls.__name__} with {name}, {sex}, {sum(nums)}')
        return cls(name, sex, sum(nums),)
    
    @staticmethod 
    def my_static_method(num1, num2):                   # Does not have special 1st agrument
        print(f'my_static_method with num1:{num1}, num2:{num2}')
        return num1 + num2


# classmethod sum_nums, which creates cls PlayerCharacter assigns to and instantiates object player3
player3 = PlayerCharacter.my_class_method('Teddy', 16, 7, 'male')

# calls run method
player3.run('Player3')

# calls staticmethod sum_num2
player3.my_static_method(16, 9)
```

```python
class PlayerCharacter:
    # class atttributes are static
    membership = False # True

    # method __init__  is the class constructor
    # self refers to the class
    def __init__(self, name, age=15):
        if (self.membership and age >= 18 ):
            # within the method self is a dynamic attributes
            self.name = name           
            self.age = age
        else:
            self.name = name
            self.age = age
            self.membership = False
    
    def run(self, player):
        print(f'{player}: {self.name} {self.age} {self.sex} min age membership {self.membership}')
    
    # classmethod decorator is bound to the class and not an instance of the class  
    @classmethod
    def sum_nums(cls, num1, num2, lvl='obj'):    # cls refers t othe the class
        if (lvl == 'class'):
            cls.membership = True
        else:
          cls.membership = False     
        nums = [num1, num2]
        print(f'@classmethod level: {lvl} membership: {cls.membership} summing {num1} + {num2} = {sum(nums)}')

# classmethod works without instantiating class 
PlayerCharacter.sum_nums(4, 5, 'class')

# instantiate
player1 = PlayerCharacter('Cindy', 15)
player2 = PlayerCharacter('Tom', 43)

# adding new attribute
player1.sex = 'female'
player2.sex = 'male'

# calling run method
player1.run('Player1')
player2.run('Player2')

# classmethod also works on instntiated class
player2.sum_nums(2, 3)
```
